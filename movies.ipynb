```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline 
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
#importing and reading the data 
 
df = pd.read_csv('movies.csv')

pd.set_option('display.max_rows', 6000)
pd.set_option('display.max_columns', 6000)
```

```python
#Looking at the data 
 
df.shape
```

```python
df.head()
```

```python
#checking for missing values

df.isna().sum()
```

```python
#creating a new dataframe without any null values
df1 = df.dropna()
df1.isna().sum()
```

```python
df1.shape
```

```python
#changing data type for the budget and gross columns 
df1.dtypes
```

```python
df1['budget'] = df1['budget'].astype('int64')

df1['gross'] = df1['gross'].astype('int64')
```

```python
df1
```

```python
#sorting the dataframe by the gross amount made 
df1 = df1.sort_values(by = 'gross', inplace = False, ascending = False)
```

```python
# dropping duplicates 

df1.drop_duplicates()
```

```python
# budget high correlation 
# company high correlation 
```

```python
#scatter plot with budget vs gross

plt.scatter(x=df1['budget'], y=df1['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film ')
```

```python
# plot the budget vs gross using seaborn 

from turtle import color

sns.regplot(x = 'budget', y = 'gross', data = df1, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# let's start looking at correlation 
df1.corr(method = 'pearson') # pearson, kendall, spearman
```

```python
# high correlation btw budget and gross
```

```python
correlation_matrix = df1.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movies Features')

plt.ylabel('Movies Features ')
```

```python
# Taking a look at the Company column 

df1.head()
```

```python
df1_numerized = df1

for col_name in df1_numerized.columns:
    if(df1_numerized[col_name].dtype == 'object'):
        df1_numerized[col_name] = df1_numerized[col_name].astype('category')
        df1_numerized[col_name] = df1_numerized[col_name].cat.codes

df1_numerized


```

```python
correlation_matrix = df1_numerized.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movies Features')

plt.ylabel('Movies Features ')
```

```python
correlation_mat = df1_numerized.corr()

corr_pairs = correlation_mat.unstack()
```

```python
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values(ascending = False)
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
```

```python
high_corr
```
